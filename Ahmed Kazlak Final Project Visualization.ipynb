```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('data1_edit.csv')
```

```python
df=data.copy()
```

```python
df.head()
```

```python
top_4_cities = df.groupby('CITY')['SALARY'].sum().nlargest(4)
bar_labels = ['red', 'blue', 'orange', 'green']
bar_colors = ['tab:red', 'tab:blue', 'tab:orange', 'tab:green']

plt.bar(x=top_4_cities.index,height=top_4_cities.values,label=bar_labels,color=bar_colors) 

plt.xticks(rotation=45)
plt.xlabel('CITY')
plt.ylabel('SALARY')
plt.title('Top 4 cities')
plt.show()
```

```python
salary_by_year = df.groupby('YEAR')['SALARY'].sum()

plt.pie(salary_by_year,
        labels = salary_by_year.index,
        autopct='%1.1f%%'
        )
plt.title('Salary by year')
plt.show()
```

```python
plt.hist(df['MARITAL_STATUS'], bins=50) 

plt.xlabel('MARITAL_STATUS') 
plt.ylabel('SALARY') 
plt.title('Salary by Marital Status') 
plt.show()
```

```python
plt.figure(figsize=(5, 3)) 
import seaborn as sns

sns.boxplot(x='MARITAL_STATUS', y='AGE', data=df) 
plt.xticks(rotation=45) 

plt.title('Age by Marital Status') 

plt.show()
```

```python
plt.figure(figsize=(5, 3))

sns.countplot(x='EDUCATION', data=df)  # univarite
plt.xticks(rotation=90)
plt.xlabel('EDUCATION') 
plt.ylabel('SALARY') 
plt.title('Sales by education') 
plt.show()
```

```python
socialtype_salary = df.groupby('MARITAL_STATUS')['SALARY'].sum()

plt.pie(socialtype_salary,
        labels=socialtype_salary.index,
        autopct='%1.1f%%', 
        wedgeprops=dict(width=0.6)
               ) 

plt.title('Salary by Marital Status') 
plt.axis('equal')
plt.show()
```
